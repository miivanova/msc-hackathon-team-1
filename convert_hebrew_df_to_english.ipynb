This notebook can be used to map the dataset as taken from the israeli website and map it to an english translation.

```python
import pandas as pd
```

```python
covid_0220_df = pd.read_csv('./data/corona_tested_individuals_ver_00220.csv')
covid_0220_df
```

```python
corona_result_dict = {
    'שלילי':'negative',
    'חיובי':'positive',
    'אחר':'other'
}

gender_dict = {
    'נקבה':'female',
    'זכר':'male'
}
```

```python
covid_0220_df['gender'] = covid_0220_df['gender'].map(gender_dict)
covid_0220_df['corona_result'] = covid_0220_df['corona_result'].map(corona_result_dict)
```

```python
covid_0220_df.to_csv('./data/corona_tested_individuals_ver_00220.english.csv', index = False)
```

```python
# Lets see how the data is changing over time

# First in terms of number of cases both negative and positive over time.


grouped_by_date_count = covid_0220_df.groupby(['test_date', 'corona_result']).count()
grouped_by_date_mean = covid_0220_df.groupby(['test_date', 'corona_result']).mean()
```
